```python
import pandas as pd
import ast

def parse_genres(x):
    try:
        return [d['name'] for d in ast.literal_eval(x)] if isinstance(x, str) else []
    except (ValueError, SyntaxError):
        return []

def load_and_clean_movies(path):
    movies = pd.read_csv(path, low_memory=False)
    for col in ['budget', 'revenue', 'popularity', 'id']:
        movies[col] = pd.to_numeric(movies[col], errors='coerce')
    movies['release_date']  = pd.to_datetime(movies['release_date'], errors='coerce')
    movies['release_year']  = movies['release_date'].dt.year
    movies['release_month'] = movies['release_date'].dt.month
    movies['genre_list']    = movies['genres'].apply(parse_genres)
    return movies

movies = load_and_clean_movies('/content/data/movies_metadata.csv')
print("shape:", movies.shape)
fin = movies[(movies['budget'] > 0) & (movies['revenue'] > 0)].copy()
fin['roi'] = fin['revenue'] / fin['budget']
print(len(fin), "films usable for ROI")
```

```python
import pandas as pd

# Dedup movies on TMDB id (movies_metadata has a few duplicate rows) — demonstrates dedup
before = len(movies)
movies = movies.drop_duplicates(subset='id')
print(f"deduped movies: {before} -> {len(movies)} (removed {before - len(movies)})")

# IMDB official ratings: tconst, averageRating, numVotes (one row per title, already unique)
imdb = pd.read_csv('/content/data/title.ratings.tsv.gz', sep='\t')
print("imdb ratings shape:", imdb.shape)

# Direct join: movies.imdb_id ('tt0114709') == imdb.tconst ('tt0114709')
movies_imdb = movies.merge(imdb, left_on='imdb_id', right_on='tconst', how='left')

print("after join:", movies_imdb.shape)
print("movies with an IMDB rating:", movies_imdb['averageRating'].notna().sum())
print(movies_imdb[['title', 'vote_average', 'averageRating', 'numVotes']].head())
```

```python
import pandas as pd

recent = pd.read_csv('/content/data_recent/TMDB_movie_dataset_v11.csv', low_memory=False)
print("asaniczka raw:", recent.shape)

# Coerce numerics
for col in ['budget', 'revenue', 'popularity', 'id', 'vote_average', 'vote_count']:
    recent[col] = pd.to_numeric(recent[col], errors='coerce')

# Dates -> year/month
recent['release_date']  = pd.to_datetime(recent['release_date'], errors='coerce')
recent['release_year']  = recent['release_date'].dt.year
recent['release_month'] = recent['release_date'].dt.month

# NOTE: asaniczka 'genres' is a COMMA string ("Drama, Action"), not dicts like Rounak Banik
def parse_genres_csv(x):
    return [g.strip() for g in x.split(',')] if isinstance(x, str) and x.strip() else []
recent['genre_list'] = recent['genres'].apply(parse_genres_csv)

# Keep only NEW films: post-2017 AND not already in our set (dedup on id)
existing_ids = set(movies_imdb['id'].dropna())
new_films = recent[(recent['release_year'] > 2017) & (~recent['id'].isin(existing_ids))].copy()
print("new post-2017 films to add:", len(new_films))

del recent   # free ~2-3GB RAM

# Merge: union columns (cast/IMDB cols become NaN for new films — expected)
combined = pd.concat([movies_imdb, new_films], ignore_index=True)
combined = combined.drop_duplicates(subset='id')
print("combined shape:", combined.shape)

# Refresh the ROI subset on the bigger table
fin = combined[(combined['budget'] > 0) & (combined['revenue'] > 0)].copy()
fin['roi'] = fin['revenue'] / fin['budget']
print("ROI films now:", len(fin), "(was 5381)")
```

```python
!pip install -q duckdb
import os, duckdb

os.makedirs('/content/drive/MyDrive/greenlight',exist_ok = True)
DB_PATH = '/content/drive/MyDrive/greenlight/greenlight.duckdb'
con = duckdb.connect(DB_PATH)

# movies table: everything EXCEPT the python-list column (DuckDB prefers scalar cols)
movies_df = combined.drop(columns=['genre_list'])

# movie_genres: one row per (movie, genre) — the clean way to do genre SQL
mg_df = combined[['id', 'genre_list']].explode('genre_list').dropna(subset=['genre_list'])
mg_df = mg_df.rename(columns={'genre_list': 'genre'})

con.execute("CREATE OR REPLACE TABLE movies       AS SELECT * FROM movies_df")
con.execute("CREATE OR REPLACE TABLE movie_genres AS SELECT * FROM mg_df")

# fin as a VIEW — auto-derives the ROI subset, no duplicate storage
con.execute("""
CREATE OR REPLACE VIEW fin AS
SELECT *, revenue / budget AS roi
FROM movies
WHERE budget > 0 AND revenue > 0
""")

con.execute("CHECKPOINT")  # flush to disk

# verify
print("movies      :", con.execute("SELECT COUNT(*) FROM movies").fetchone()[0])
print("movie_genres:", con.execute("SELECT COUNT(*) FROM movie_genres").fetchone()[0])
print("fin (ROI)   :", con.execute("SELECT COUNT(*) FROM fin").fetchone()[0])
con.close()
print("saved to", DB_PATH)
```
