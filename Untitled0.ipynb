````python
# ============================================================
#  GREENLIGHT — ONE-CELL SETUP   (run once per session)
# ============================================================
import os, json, subprocess, time, requests
from google.colab import drive

# 1. Mount Drive (plain, safe)
drive.mount('/content/drive')

# 2. Start Ollama (install only if missing), model from Drive
os.environ["OLLAMA_MODELS"] = "/content/drive/MyDrive/greenlight/ollama_models"
if subprocess.run(["which","ollama"], capture_output=True).returncode != 0:
    print("installing ollama..."); os.system("apt-get -qq install -y zstd")
    os.system("curl -fsSL https://ollama.com/install.sh | sh")
os.system("pkill -f 'ollama serve'"); time.sleep(2)
subprocess.Popen(["ollama","serve"], stdout=open("/tmp/ollama.log","w"), stderr=subprocess.STDOUT)
for _ in range(30):
    try:
        m=[x["name"] for x in requests.get("http://localhost:11434/api/tags",timeout=2).json().get("models",[])]
        print("✅ Ollama UP. models:", m); break
    except Exception: time.sleep(2)
else: print("❌ Ollama failed"); os.system("tail -n 20 /tmp/ollama.log")

# 3. Load constraints (the data->AI bridge — no DB needed at runtime)
with open('/content/drive/MyDrive/greenlight/constraints.json') as f:
    CONSTRAINTS = json.load(f)

# 4. LLM helper
def call_llm(prompt, model="qwen2.5:7b", max_tokens=350, fmt=None):
    payload={"model":model,"prompt":prompt,"stream":False,
             "options":{"num_predict":max_tokens},"keep_alive":"10m"}
    if fmt: payload["format"]=fmt
    data=requests.post("http://localhost:11434/api/generate",json=payload).json()
    if "response" not in data: raise RuntimeError(f"Ollama error: {data}")
    return data["response"].strip()

# 5. Agents
def resolve_window(genre, c, window=None):
    if window: return window
    return ", ".join(c.get("seasonal_by_genre",{}).get(genre) or c["seasonal_fit"]["best_months_named"])

def writer_agent(genre, c, target_window, feedback=None):
    prompt=f"""You are a screenwriter for a major film studio.
Write a compelling 150-word film synopsis for a {genre} film planned for a {target_window} release.

Market guidance (from 20 years of box-office data):
- Release window: {target_window}. Bake this season into the story's SETTING, plot hooks, and tone.
- Maximize broad audience appeal. {c["audience_signal"]["note"]}

Write the entire synopsis in ENGLISH only.
Output ONLY a title and the ~150-word synopsis. No extra commentary."""
    if feedback: prompt+=f"\n\nRevise to fix these issues from the previous draft:\n{feedback}"
    return call_llm(prompt)

def critic_agent(synopsis, genre, c, target_window):
    prompt=f"""You are a strict film market analyst. Score a synopsis against box-office data.

SYNOPSIS:
{synopsis}

REQUIREMENTS:
1. Genre is "{genre}" — must clearly read as this genre.
2. Targets a {target_window} release — story's season/setting should fit.
3. Broad audience appeal.
4. Coherent and well-structured.

Score 0.0-1.0 (be critical). Return ONLY this JSON:
{{"score": <float>, "passed": [<strings>], "failed": [<strings>], "suggestions": [<strings>]}}"""
    raw=call_llm(prompt, max_tokens=400, fmt="json")
    try: return json.loads(raw)
    except json.JSONDecodeError:
        return {"score":0.0,"passed":[],"failed":["invalid JSON"],"suggestions":["retry"]}

def refiner_agent(genre, c, target_window, critique):
    fb="Fix these issues from the previous draft:\n"+"\n".join(f"- {x}" for x in critique.get("failed",[]))
    if critique.get("suggestions"): fb+="\nSuggestions:\n"+"\n".join(f"- {x}" for x in critique["suggestions"])
    return writer_agent(genre, c, target_window, feedback=fb)

# 6. LangGraph loop
os.system("pip install -q langgraph")
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, List
class GenState(TypedDict):
    genre:str; constraints:dict; window:str; target_window:str
    synopsis:str; score:float; critique:dict
    iteration:int; history:List[dict]; best_synopsis:str; best_score:float
def writer_node(s):
    t=resolve_window(s["genre"],s["constraints"],s.get("window"))
    return {"synopsis":writer_agent(s["genre"],s["constraints"],t),"target_window":t,"iteration":1}
def critic_node(s):
    r=critic_agent(s["synopsis"],s["genre"],s["constraints"],s["target_window"])
    print(f"  [iter {s['iteration']}] window={s['target_window']} score={r['score']} failed={r.get('failed')}")
    u={"score":r["score"],"critique":r,"history":s["history"]+[{"iteration":s["iteration"],"score":r["score"]}]}
    if r["score"]>s["best_score"]: u["best_synopsis"],u["best_score"]=s["synopsis"],r["score"]
    return u
def refiner_node(s):
    return {"synopsis":refiner_agent(s["genre"],s["constraints"],s["target_window"],s["critique"]),
            "iteration":s["iteration"]+1}
def should_continue(s):
    if s["score"]>0.7: return "end"
    if s["iteration"]>=3: return "end"
    if len(s["history"])>=2 and s["history"][-1]["score"]<=s["history"][-2]["score"]: return "end"
    return "refiner"
g=StateGraph(GenState)
for n,f in [("writer",writer_node),("critic",critic_node),("refiner",refiner_node)]: g.add_node(n,f)
g.add_edge(START,"writer"); g.add_edge("writer","critic")
g.add_conditional_edges("critic",should_continue,{"refiner":"refiner","end":END}); g.add_edge("refiner","critic")
app=g.compile()

# 7. One-liner
def generate_synopsis(genre, window=None):
    r=app.invoke({"genre":genre,"constraints":CONSTRAINTS,"window":window,"target_window":"",
        "synopsis":"","score":0.0,"critique":{},"iteration":0,"history":[],"best_synopsis":"","best_score":0.0})
    return {"genre":genre,"window":r["target_window"],"synopsis":r["best_synopsis"],"score":r["best_score"],
            "iterations":r["iteration"],"history":[h["score"] for h in r["history"]],"critique":r["critique"]}

print("✅ SETUP COMPLETE — now run: generate_synopsis('Horror')")
````

Clean + combine + build ALL tables in a LOCAL DuckDB:

````python
import pandas as pd, ast, duckdb, os

def parse_genres(x):
    try: return [d['name'] for d in ast.literal_eval(x)] if isinstance(x,str) else []
    except (ValueError, SyntaxError): return []

# --- Rounak Banik movies ---
movies = pd.read_csv('/content/data/movies_metadata.csv', low_memory=False)
for c in ['budget','revenue','popularity','id']:
    movies[c] = pd.to_numeric(movies[c], errors='coerce')
movies['release_date']  = pd.to_datetime(movies['release_date'], errors='coerce')
movies['release_year']  = movies['release_date'].dt.year
movies['release_month'] = movies['release_date'].dt.month
movies['genre_list']    = movies['genres'].apply(parse_genres)
movies = movies.drop_duplicates(subset='id')

# --- IMDB ratings join ---
imdb = pd.read_csv('/content/data/title.ratings.tsv.gz', sep='\t')
movies_imdb = movies.merge(imdb, left_on='imdb_id', right_on='tconst', how='left')

# --- asaniczka recent (post-2017, dedup on id) ---
recent = pd.read_csv('/content/data_recent/TMDB_movie_dataset_v11.csv', low_memory=False)
for c in ['budget','revenue','popularity','id','vote_average','vote_count']:
    recent[c] = pd.to_numeric(recent[c], errors='coerce')
recent['release_date']  = pd.to_datetime(recent['release_date'], errors='coerce')
recent['release_year']  = recent['release_date'].dt.year
recent['release_month'] = recent['release_date'].dt.month
recent['genre_list']    = recent['genres'].apply(lambda x: [g.strip() for g in x.split(',')] if isinstance(x,str) and x.strip() else [])
existing = set(movies_imdb['id'].dropna())
new_films = recent[(recent['release_year']>2017) & (~recent['id'].isin(existing))].copy()
combined = pd.concat([movies_imdb, new_films], ignore_index=True).drop_duplicates(subset='id')
del movies, movies_imdb, recent, new_films, imdb   # free RAM

# --- credits: directors + top 3 cast ---
credits = pd.read_csv('/content/data/credits.csv')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')
def get_director(s):
    try:
        crew=ast.literal_eval(s); n=[c['name'] for c in crew if c.get('job')=='Director']; return n[0] if n else None
    except (ValueError,SyntaxError): return None
def get_top_cast(s, n=3):
    try: return [c['name'] for c in ast.literal_eval(s)[:n]]
    except (ValueError,SyntaxError): return []
credits['director'] = credits['crew'].apply(get_director)
credits['top_cast'] = credits['cast'].apply(get_top_cast)
directors_df = credits[['id','director']].dropna(subset=['director'])
actors_df    = credits[['id','top_cast']].explode('top_cast').dropna(subset=['top_cast']).rename(columns={'top_cast':'actor'})

# --- write to LOCAL DuckDB (never Drive directly!) ---
LOCAL_DB = '/content/greenlight.duckdb'
if os.path.exists(LOCAL_DB): os.remove(LOCAL_DB)
con = duckdb.connect(LOCAL_DB)
movies_df = combined.drop(columns=['genre_list'])
mg_df = combined[['id','genre_list']].explode('genre_list').dropna(subset=['genre_list']).rename(columns={'genre_list':'genre'})
con.execute("CREATE TABLE movies       AS SELECT * FROM movies_df")
con.execute("CREATE TABLE movie_genres AS SELECT * FROM mg_df")
con.execute("CREATE TABLE directors    AS SELECT * FROM directors_df")
con.execute("CREATE TABLE movie_actors AS SELECT * FROM actors_df")
con.execute("""CREATE TABLE movielens AS
SELECT CAST(l.tmdbId AS BIGINT) AS id, ROUND(AVG(r.rating),2) AS ml_avg_rating, COUNT(*) AS ml_n_ratings
FROM read_csv_auto('/content/data/ratings.csv') r
JOIN read_csv_auto('/content/data/links.csv') l ON r.movieId=l.movieId
WHERE l.tmdbId IS NOT NULL GROUP BY l.tmdbId""")
con.execute("""CREATE OR REPLACE VIEW fin AS
SELECT *, revenue/budget AS roi FROM movies WHERE budget>=100000 AND revenue>=1000""")
con.execute("CHECKPOINT")
for t in ['movies','movie_genres','directors','movie_actors','movielens']:
    print(f"  {t:14}", con.execute(f"SELECT COUNT(*) FROM {t}").fetchone()[0])
print("  fin (ROI)     ", con.execute("SELECT COUNT(*) FROM fin").fetchone()[0])
print("✅ local DB built")
````

Regenerate constraints.json (REAL director/actor + per-genre seasonal, one clean builder):

````python
import json, datetime
month_names = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}

genres = con.execute("""SELECT g.genre, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN movie_genres g ON f.id=g.id GROUP BY g.genre HAVING COUNT(*)>=30 ORDER BY median_roi DESC LIMIT 5""").df()
season = con.execute("""SELECT release_month, COUNT(*) AS n_films, ROUND(MEDIAN(roi),2) AS median_roi
FROM fin WHERE release_month IS NOT NULL GROUP BY release_month ORDER BY median_roi DESC""").df()
directors = con.execute("""SELECT d.director, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN directors d ON f.id=d.id GROUP BY d.director HAVING COUNT(*)>=5 ORDER BY median_roi DESC LIMIT 10""").df()
actors = con.execute("""SELECT a.actor, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN movie_actors a ON f.id=a.id GROUP BY a.actor HAVING COUNT(*)>=5 ORDER BY median_roi DESC LIMIT 10""").df()
aud = con.execute("""SELECT CASE WHEN ml.ml_avg_rating>=3.8 THEN 'loved' WHEN ml.ml_avg_rating>=3.2 THEN 'liked' ELSE 'mixed' END AS band,
ROUND(MEDIAN(f.roi),2) AS median_roi FROM fin f JOIN movielens ml ON f.id=ml.id WHERE ml.ml_n_ratings>=50 GROUP BY band""").df().set_index('band')['median_roi'].to_dict()
sbg = con.execute("""SELECT g.genre, f.release_month AS month, COUNT(*) AS n, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN movie_genres g ON f.id=g.id WHERE f.release_month IS NOT NULL GROUP BY g.genre, f.release_month""").df()

genre_windows = {}
for g in sbg['genre'].unique():
    top = sbg[(sbg['genre']==g) & (sbg['n']>=15)].sort_values('median_roi', ascending=False).head(3)
    if len(top): genre_windows[g] = [month_names[int(m)] for m in top['month']]

best  = season.head(3)['release_month'].astype(int).tolist()
worst = season.tail(2)['release_month'].astype(int).tolist()
CONSTRAINTS = {
  "meta": {"generated_on": datetime.date.today().isoformat(),
           "source":"TMDB (Rounak Banik + asaniczka) + IMDB ratings",
           "n_roi_films": int(con.execute("SELECT COUNT(*) FROM fin").fetchone()[0]),
           "budget_floor_usd":100000, "roi_metric":"median of revenue/budget"},
  "top_genres": json.loads(genres.to_json(orient="records")),
  "seasonal_fit": {"best_months":best, "best_months_named":[month_names[m] for m in best],
                   "worst_months_named":[month_names[m] for m in worst],
                   "note": f"{month_names[best[0]]}/{month_names[best[1]]} and {month_names[best[2]]} strongest overall."},
  "seasonal_by_genre": genre_windows,
  "director_trends": json.loads(directors.to_json(orient="records")),
  "actor_trends": json.loads(actors.to_json(orient="records")),
  "audience_signal": {"loved_roi": aud.get("loved"), "mixed_roi": aud.get("mixed"),
                      "note":"Films audiences love (MovieLens >=3.8) earn ~2.5x the ROI of mixed films. Prioritize broad appeal."}
}
with open('/content/constraints.json','w') as f:
    json.dump(CONSTRAINTS, f, indent=2, ensure_ascii=False)
print("✅ constraints built. keys:", list(CONSTRAINTS.keys()))
print("Horror:", genre_windows.get("Horror"), "| Action:", genre_windows.get("Action"))
print("directors:", [d['director'] for d in CONSTRAINTS['director_trends'][:3]])
````

````python
# ============================================================
#  PHASE 8 — DETERMINISTIC, TESTED CRITIC (validator)
# ============================================================
import re

GENRE_LEXICON = {
    "horror":     {"horror","terror","fear","haunt","nightmare","dread","evil","demon","ghost","blood","dark","death","sinister","supernatural","scream","monster","curse","possessed","killer","shadow","grim"},
    "action":     {"action","explosive","fight","chase","battle","mission","weapon","danger","combat","hero","escape","showdown","relentless","adrenaline","pursuit","gun","war","strike","ruthless"},
    "comedy":     {"comedy","hilarious","laugh","funny","witty","absurd","misadventure","awkward","chaos","prank","goofy","humor","quirky","mishap"},
    "drama":      {"drama","emotional","struggle","grief","family","loss","redemption","heart","journey","conflict","betrayal","sacrifice","truth","haunted"},
    "romance":    {"romance","love","heart","passion","relationship","longing","desire","affair","wedding","soulmate","tender","fall"},
    "thriller":   {"thriller","suspense","tension","danger","secret","conspiracy","hunt","twist","deadly","betrayal","race","mystery","stalk"},
    "science fiction": {"future","space","alien","technology","robot","dystopia","time","galaxy","cyber","experiment","planet","artificial"},
    "scifi":      {"future","space","alien","technology","robot","dystopia","time","galaxy","cyber","experiment","planet","artificial"},
    "fantasy":    {"magic","kingdom","dragon","quest","myth","sorcerer","realm","enchanted","prophecy","legend","wizard","spell"},
    "animation":  {"adventure","friendship","colorful","family","journey","wonder","heart","magical","young","discover"},
    "family":     {"family","friendship","heartwarming","adventure","together","young","wholesome","journey","home","love"},
    "mystery":    {"mystery","clue","detective","secret","murder","investigate","puzzle","hidden","truth","suspect","unravel"},
    "crime":      {"crime","heist","gang","detective","murder","police","criminal","robbery","underworld","justice","mob"},
}
CRITICAL = {"genre_signal", "window_consistent", "no_placeholder"}

def _genre_terms(genre):
    g = (genre or "").strip().lower()
    terms = set(GENRE_LEXICON.get(g, set()))
    terms.add(g)                       # the genre word itself always counts
    return terms

def _month_match(target_window, best):
    tw = (target_window or "").lower()
    return any(m.strip().lower()[:3] in tw for m in best)   # "Jan" matches "January", "Jun"↔"June"

def validate_synopsis(synopsis, genre, target_window, constraints, user_window=None, target_len = 80):
    """Pure, deterministic. (synopsis, plan) -> {score, passed, failed, suggestions, valid}."""
    s = (synopsis or "").strip()
    words = re.findall(r"[A-Za-z']+", s)
    n = len(words); low = s.lower()
    checks = []   # (name, passed, weight, suggestion)

    # 1. length (band scales with requested length; default 80 -> 40-120)
    lo, hi = round(target_len * 0.5), round(target_len * 1.5)
    checks.append(("length_ok", lo <= n <= hi, 0.20,
        f"Synopsis is {n} words; aim for ~{target_len} ({lo}-{hi})."))
    checks.append(("length_ok", 40 <= n <= 120, 0.20,
        f"Synopsis is {n} words; aim for 40-120 (logline to short paragraph)."))

    # 2. genre signal
    terms = _genre_terms(genre)
    hits = sorted({w for w in (x.lower() for x in words) if w in terms})
    checks.append(("genre_signal", len(hits) > 0, 0.25,
        f"Add {genre}-appropriate language (e.g. {', '.join(list(terms)[:5])})."))

    # 3. no placeholders / leaked formatting / meta-text
    bad = ["todo","lorem","[insert","{","}","```","as an ai","i cannot","here is","synopsis:","note:"]
    checks.append(("no_placeholder", not any(b in low for b in bad), 0.15,
        "Remove placeholders, JSON/markdown artifacts, or meta-text like 'Here is the synopsis'."))

    # 4. complete (terminal punctuation + at least 2 sentences)
    sentences = [x for x in re.split(r"[.!?]+", s) if x.strip()]
    checks.append(("complete", s.endswith((".","!","?")) and len(sentences) >= 2, 0.15,
        "End with terminal punctuation and write at least two full sentences (no truncation)."))

    # 5. window consistency (is the release plan aligned with the data?)
    best = constraints.get("seasonal_by_genre", {}).get(genre) \
           or constraints.get("seasonal_fit", {}).get("best_months_named", [])
    if user_window:
        window_ok = bool(target_window and target_window.strip())
        win_sug = "User override given; ensure target_window reflects the user's requested timing."
    else:
        window_ok = _month_match(target_window, best)
        win_sug = f"Release window should match the data-best months for {genre}: {', '.join(best)}."
    checks.append(("window_consistent", window_ok, 0.25, win_sug))

    passed      = [c[0] for c in checks if c[1]]
    failed      = [c[0] for c in checks if not c[1]]
    suggestions = [c[3] for c in checks if not c[1]]
    score       = round(sum(c[2] for c in checks if c[1]), 3)
    critical_failed = [f for f in failed if f in CRITICAL]
    valid = score >= 0.7 and not critical_failed

    return {"score": score, "passed": passed, "failed": failed,
            "suggestions": suggestions, "n_words": n, "genre_hits": hits,
            "critical_failed": critical_failed, "valid": valid}

print("✅ validator defined")
````

The langraph(cell 4)

````python
from langgraph.graph import StateGraph, END
from typing import TypedDict, List, Dict, Any
import re

_CUR_STAGE = "Writer"
def _set_stage(s):
    global _CUR_STAGE
    _CUR_STAGE = s

# --- language guard: qwen sometimes drifts into Chinese on long outputs ---
_CJK = re.compile(r'[\u3000-\u9fff\uff00-\uffef\u3040-\u30ff\uac00-\ud7af]')
def _en(text):
    """Guarantee English. If CJK appears, rewrite to English; last resort strip it."""
    if not text or not _CJK.search(text):
        return text
    try:
        fixed = call_llm("Rewrite the following film synopsis ENTIRELY in fluent English, keeping the "
                         "same story, characters, tone, and length. Output ONLY the English prose, "
                         "nothing else:\n\n" + text, max_tokens=900)
        if fixed and not _CJK.search(fixed):
            return fixed.strip()
    except Exception:
        pass
    return _CJK.sub("", text)                      # last resort: drop any remaining CJK

class GenState(TypedDict):
    genre: str; genres: List[str]; constraints: Dict[str, Any]; window: Any; length: int
    target_window: str; premise: str; inspirations: List[str]
    synopsis: str; score: float; critique: Dict[str, Any]; iteration: int
    history: List[Dict[str, Any]]; best_synopsis: str; best_score: float

def _tok(n): return min(900, int(n * 1.8) + 120)

def writer_agent(state):
    _set_stage("Writer")
    genre = state["genre"]; c = state["constraints"]; L = state.get("length") or 80
    genres = state.get("genres") or [genre]
    insp = state.get("inspirations") or []
    tw = resolve_window(genre, c, state.get("window"))
    lo, hi = round(L * 0.5), round(L * 1.5)
    aud = c.get("audience_signal", {}).get("note", "")
    prem = (state.get("premise") or "").strip()
    premise_block = ""
    if prem:
        premise_block = ("\nUSER'S REQUIRED PREMISE — build the story around this; keep its characters "
                         f'and key events, do NOT swap in a different plot:\n"{prem}"\n')
    insp_block = ""
    if len(insp) > 1:
        insp_block = ("\nFUSE the core concepts, tone, and energy of these films into ONE original story "
                      "(do NOT copy their plots, characters, or names): " + ", ".join(insp) + ".\n")
    elif insp:
        insp_block = (f"\nTake tonal and structural inspiration from {insp[0]} "
                      "(do NOT copy its plot, characters, or names).\n")
    try: ex = rag_examples(genre, k=3)
    except Exception: ex = []
    rag_block = ""
    if ex:
        rag_block = ("\nStructural inspiration from real high-ROI films (study tone & structure; do NOT "
                     "copy plots or names):\n" + "\n".join(f"- {e[:300]}" for e in ex) + "\n")
    genre_desc = (" + ".join(genres) + " blend") if len(genres) > 1 else genre
    prompt = f"""You are a film development writer. Write an original {genre_desc} movie synopsis.
{premise_block}{insp_block}
Data-informed guidance:
- Target release window: {tw}.
- {aud}
{rag_block}
Rules:
- Write the ENTIRE synopsis in ENGLISH ONLY — never use Chinese or any non-English characters.
- About {L} words ({lo}-{hi}), vivid, and unmistakably a {genre_desc}.
- If a premise is given, the plot must follow it faithfully.
- If inspirations are given, capture their spirit but keep it fully ORIGINAL.
- Output ONLY the synopsis prose. No title, no labels, no preamble, no markdown."""
    return {"synopsis": _en(call_llm(prompt, max_tokens=_tok(L))), "target_window": tw}

def critic_agent(state):
    _set_stage("Critic")
    r = validate_synopsis(state["synopsis"], state["genre"], state["target_window"],
                          CONSTRAINTS, user_window=state.get("window"),
                          target_len=state.get("length") or 80)
    critique = {"score": r["score"], "passed": r["passed"], "failed": r["failed"],
                "suggestions": r["suggestions"], "valid": r["valid"]}
    history = state["history"] + [{"iteration": state["iteration"],
                                   "score": r["score"], "synopsis": state["synopsis"]}]
    best_s, best_sc = state["best_synopsis"], state["best_score"]
    if r["score"] > best_sc: best_sc, best_s = r["score"], state["synopsis"]
    return {"score": r["score"], "critique": critique, "history": history,
            "best_synopsis": best_s, "best_score": best_sc}

def refiner_agent(state):
    _set_stage("Refiner")
    L = state.get("length") or 80
    prem = (state.get("premise") or "").strip()
    premise_block = f'\nStay faithful to the user\'s premise: "{prem}"\n' if prem else ""
    fixes = "\n".join(f"- {s}" for s in state["critique"].get("suggestions", [])) or "- Tighten and polish."
    prompt = f"""You are a script doctor. Rewrite this {state['genre']} film synopsis to fix the issues.

Current synopsis:
{state['synopsis']}
{premise_block}
Fix these specific problems:
{fixes}

Rules:
- Write the ENTIRE synopsis in ENGLISH ONLY — never use Chinese or any non-English characters.
- About {L} words, vivid, clearly {state['genre']}.
- Planned for release in: {state['target_window']}.
- Do not drift from the user's premise (if any).
- Output ONLY the revised synopsis prose. No labels, no preamble, no markdown."""
    return {"synopsis": _en(call_llm(prompt, max_tokens=_tok(L))), "iteration": state["iteration"] + 1}

def should_continue(state):
    if state["score"] > 0.7:    return "end"
    if state["iteration"] >= 3: return "end"
    h = state["history"]
    if len(h) >= 2 and h[-1]["score"] <= h[-2]["score"]: return "end"
    return "refine"

g = StateGraph(GenState)
g.add_node("writer", writer_agent); g.add_node("critic", critic_agent); g.add_node("refiner", refiner_agent)
g.set_entry_point("writer")
g.add_edge("writer", "critic")
g.add_conditional_edges("critic", should_continue, {"refine": "refiner", "end": END})
g.add_edge("refiner", "critic")
app = g.compile()

def generate_synopsis(genre, window=None, length=80, premise=None, genres=None, inspirations=None):
    r = app.invoke({"genre": genre, "genres": genres or [genre], "constraints": CONSTRAINTS,
                    "window": window, "length": length, "target_window": "",
                    "premise": premise or "", "inspirations": inspirations or [],
                    "synopsis": "", "score": 0.0, "critique": {}, "iteration": 0,
                    "history": [], "best_synopsis": "", "best_score": 0.0})
    return {"genre": genre, "window": r["target_window"], "synopsis": r["best_synopsis"],
            "score": r["best_score"], "iterations": r["iteration"],
            "history": [h["score"] for h in r["history"]], "critique": r["critique"]}

print("✅ graph: per-agent progress + English-only guard")
````

Parser (Cell 5)

````python
import json, re

KNOWN_GENRES = ["Action","Adventure","Animation","Comedy","Crime","Documentary","Drama",
                "Family","Fantasy","History","Horror","Music","Mystery","Romance",
                "Science Fiction","Thriller","War","Western"]
SEASON_TO_MONTH = {"summer":"June","winter":"December","spring":"April","fall":"October",
                   "autumn":"October","holiday":"December","christmas":"December","halloween":"October"}
MONTHS = ["January","February","March","April","May","June","July","August",
          "September","October","November","December"]

def _match_genre(text):
    t = (text or "").lower()
    for g in KNOWN_GENRES:
        if g.lower() in t: return g
    if any(k in t for k in ("sci-fi","scifi","sci fi")): return "Science Fiction"
    if any(k in t for k in ("rom-com","romcom","rom com")): return "Romance"
    return None

def _match_genres(text):
    t = (text or "").lower()
    found = [g for g in KNOWN_GENRES if g.lower() in t]
    if any(k in t for k in ("sci-fi","scifi","sci fi")) and "Science Fiction" not in found:
        found.append("Science Fiction")
    if any(k in t for k in ("rom-com","romcom","rom com")):
        for g in ("Romance","Comedy"):
            if g not in found: found.append(g)
    return found

_PREMISE_TRIGGERS = ("where ","about ","in which ","story of ","involving ","featuring ",
                     "following ","centered on ","centred on ","premise:","plot:")
def _premise_fallback(text):
    t = (text or "").strip(); low = t.lower(); best, bestpos = None, len(t) + 1
    for k in _PREMISE_TRIGGERS:
        i = low.find(k)
        if i != -1 and i < bestpos: bestpos, best = i, t[i+len(k):].strip(" .,:")
    return best or None

_INSP_TRIGGERS = ("take inspiration from","inspiration from","inspired by","based on",
                  "in the style of","combining the movies","combining","combine the movies",
                  "combine","mix of","fusion of","cross between","mashup of","blend of",
                  "like the movies","like the movie","like the films","like the film")
_INSP_PREFIX = re.compile(r"^(?:the\s+)?(?:movies?|films?)\s+", re.I)
def _inspirations_fallback(text):
    low = (text or "").lower()
    for k in _INSP_TRIGGERS:
        i = low.find(k)
        if i != -1:
            tail = _INSP_PREFIX.sub("", text[i+len(k):].strip(" :"))
            parts = re.split(r"\s*(?:,|&|\band\b|\bplus\b|\bwith\b)\s*", tail, flags=re.I)
            titles = [p.strip(" .\"'") for p in parts if p.strip()]
            titles = [t for t in titles if 1 <= len(t) <= 40]
            return titles[:3] or None
    return None

def _parse_budget(text):
    t = (text or "").lower().replace(",", "")
    mults = {"billion":1e9,"b":1e9,"million":1e6,"m":1e6,"mil":1e6,"thousand":1e3,"k":1e3}
    m = (re.search(r"\$\s*(\d+(?:\.\d+)?)\s*(billion|million|thousand|mil|b|m|k)?", t)
         or re.search(r"(\d+(?:\.\d+)?)\s*(billion|million|thousand|mil|b|m|k)\b\s*\$?", t)
         or re.search(r"budget\D{0,12}(\d+(?:\.\d+)?)\s*(billion|million|thousand|mil|b|m|k)?", t))
    if not m: return None
    val = float(m.group(1)); mult = mults.get(m.group(2) or "", 1)
    if mult == 1 and val < 1000: mult = 1e6
    return int(val * mult)

def _normalize(data, text):
    genre = data.get("genre")
    if genre not in KNOWN_GENRES:
        genre = _match_genre(text) or "Drama"
    genres = [g for g in (data.get("genres") or []) if g in KNOWN_GENRES]
    for g in _match_genres(text):
        if g not in genres: genres.append(g)
    if not genres: genres = [genre]
    genre = genres[0]
    window = data.get("window")
    if isinstance(window, str):
        w = window.strip().lower()
        if w in ("","null","none","best","any","data-driven"): window = None
        elif w in SEASON_TO_MONTH: window = SEASON_TO_MONTH[w]
        else: window = next((mo for mo in MONTHS if mo.lower()[:3] in w), None)
    else: window = None
    length = data.get("length")
    lm = re.search(r"(\d{2,4})\s*word", (text or "").lower())
    if lm: length = int(lm.group(1))
    if not isinstance(length, int) or not (30 <= length <= 600): length = 80
    budget = _parse_budget(text)
    premise = data.get("premise")
    if isinstance(premise, str):
        premise = premise.strip()
        if premise.lower() in ("","null","none","n/a"): premise = None
    else: premise = None
    if not premise: premise = _premise_fallback(text)
    inspirations = data.get("inspirations")
    if isinstance(inspirations, list):
        inspirations = [str(x).strip(" .\"'") for x in inspirations if str(x).strip()]
    else: inspirations = []
    if not inspirations: inspirations = _inspirations_fallback(text) or []
    return {"genre": genre, "genres": genres, "window": window, "length": length,
            "budget": budget, "premise": premise, "inspirations": inspirations}

def parse_prompt(text):
    schema = ('{"genre":"<primary genre>","genres":["<all genres named>"],'
              '"window":"<month/season, or null>","length":<int or null>,'
              '"budget":<int dollars or null>,'
              '"premise":"<specific plot the user describes, or null>",'
              '"inspirations":["<movies to take inspiration from / combine, or empty>"]}')
    prompt = (f"Extract fields from this film request. Reply ONLY JSON: {schema}\n"
              f"Allowed genres: {', '.join(KNOWN_GENRES)}.\n"
              f"If they say 'best release window' or give no timing, window=null.\n"
              f"Put any specific story they describe into premise; else premise=null.\n"
              f"List EVERY genre they name in genres ('horror comedy' -> both). "
              f"Put ONLY movies the user EXPLICITLY names (as inspiration or to combine) into inspirations "
              f'(\"combine Matrix and Inception\" -> ["The Matrix","Inception"]); if they name no movie, '
              f"inspirations=[]. Do NOT infer or guess movies from the plot.\n"


              f"Request: {text}")
    try: data = json.loads(call_llm(prompt, max_tokens=220, fmt="json"))
    except Exception: data = {}
    return _normalize(data, text)

def generate_from_prompt(text):
    p = parse_prompt(text); print("🧭 parsed:", p)
    return generate_synopsis(p["genre"], window=p["window"], length=p["length"],
                             premise=p.get("premise"), genres=p.get("genres"),
                             inspirations=p.get("inspirations"))

print("✅ parser ready (multi-genre + movie inspirations)")
````

````python
# ============================================================
#  CAST-BY-GENRE-BUDGET — who returns ROI in each genre at each budget tier (CPU only)
# ============================================================
import duckdb, shutil, json
from collections import defaultdict

DRIVE_DB="/content/drive/MyDrive/greenlight/greenlight.duckdb"; LOCAL_DB="/content/greenlight.duckdb"
shutil.copy(DRIVE_DB, LOCAL_DB)
con=duckdb.connect(LOCAL_DB)
con.execute("CREATE OR REPLACE VIEW fin AS SELECT *, revenue/budget AS roi "
            "FROM movies WHERE budget>=100000 AND revenue>=1000")

def col(tbl, want, fb):
    cols=[r[0] for r in con.execute(f"DESCRIBE {tbl}").fetchall()]
    for w in want:
        for c in cols:
            if w in c.lower(): return c
    return fb
g_col=col("movie_genres",["genre"],"genre"); g_key=col("movie_genres",["id"],"id")
a_col=col("movie_actors",["actor","name"],"actor"); a_key=col("movie_actors",["id"],"id")
d_col=col("directors",["director","name"],"director"); d_key=col("directors",["id"],"id")
print("cols ->", g_col, a_col, d_col)

TIER="CASE WHEN f.budget<1e7 THEN 'lean' WHEN f.budget<1e8 THEN 'mid' ELSE 'big' END"
def top_people(tbl, p_col, p_key, min_n):
    q=f"""SELECT mg.{g_col} genre, {TIER} tier, p.{p_col} person,
                 COUNT(*) n, ROUND(MEDIAN(f.roi),2) med
          FROM fin f
          JOIN movie_genres mg ON f.id=mg.{g_key}
          JOIN {tbl} p ON f.id=p.{p_key}
          WHERE p.{p_col} IS NOT NULL AND p.{p_col} <> ''
          GROUP BY 1,2,3 HAVING COUNT(*)>={min_n}"""
    return con.execute(q).fetchall()
actor_rows=top_people("movie_actors", a_col, a_key, 2)
dir_rows  =top_people("directors",    d_col, d_key, 2)
con.close()

def rank(rows, topn):
    d=defaultdict(list)
    for genre,tier,person,n,med in rows:
        d[(genre,tier)].append((person,int(n),float(med or 0)))
    out=defaultdict(dict)
    for (genre,tier),lst in d.items():
        lst.sort(key=lambda x:(x[1],x[2]), reverse=True)   # prolific + profitable first
        out[genre][tier]=[{"name":p,"n":n,"roi":r} for p,n,r in lst[:topn]]
    return out
actors=rank(actor_rows,6); directors=rank(dir_rows,3)

cast={}
for g in set(list(actors)+list(directors)):
    cast[g]={t:{"actors":actors.get(g,{}).get(t,[]),
                "directors":directors.get(g,{}).get(t,[])} for t in ["lean","mid","big"]}

CPATH="/content/constraints.json"
shutil.copy("/content/drive/MyDrive/greenlight/constraints.json",CPATH)
c=json.load(open(CPATH)); c["cast_by_genre_budget"]=cast
json.dump(c,open(CPATH,"w"),indent=2,ensure_ascii=False)
shutil.copy(CPATH,"/content/drive/MyDrive/greenlight/constraints.json")

print(f"\n✅ cast_by_genre_budget for {len(cast)} genres\n")
for g in ["Horror","Science Fiction","Action","Drama"]:
    if g in cast:
        for t in ["lean","mid","big"]:
            a=[x["name"] for x in cast[g][t]["actors"][:3]]
            print(f"{g:16} {t:4} -> {a}")
````

````python
# ---- /chat: explicit "replace X with Y" + gender-aware auto-recast + premise-faithful edits ----
import re as _re
from pydantic import BaseModel

SWAP_WORDS = ("replace","swap","change","instead","remove","don't","do not","dont",
              "not a fan","different","someone else","other","recast")

def _explicit_target(text):
    m = _re.search(r"(?:replace|swap|change|recast|sub(?:stitute)?)\s+(.+?)\s+(?:with|for|to|by|->|→)\s+(.+?)[.?!]*$",
                   text, _re.I)
    if m:
        return m.group(1).strip(" '\""), m.group(2).strip(" '\"")
    return None, None

def _name_targets(people, msg):
    words = set(_re.findall(r"[a-z]+", msg)); msg_ns = _re.sub(r"[^a-z]", "", msg); out = []
    for p in people:
        nm = p["name"].lower(); toks = [t for t in _re.findall(r"[a-z]+", nm) if len(t) >= 3]
        nm_ns = _re.sub(r"[^a-z]", "", nm)
        if nm in msg or (nm_ns and nm_ns in msg_ns) or (toks and toks[-1] in words):
            out.append(p)
    return out

def _people_pool(genre, budget, which):
    cbg = CONSTRAINTS.get("cast_by_genre_budget", {}).get(genre, {})
    tier = _tier_for(budget); order = [tier] + [t for t in ["big","mid","lean"] if t != tier]; pool = []
    for t in order:
        for p in cbg.get(t, {}).get(which, []):
            if p["name"] not in pool: pool.append(p["name"])
    glob = "actor_trends" if which == "actors" else "director_trends"
    fld = "actor" if which == "actors" else "director"
    for x in CONSTRAINTS.get(glob, []):
        nm = (x.get(fld) or x.get("name")) if isinstance(x, dict) else x
        if nm and nm not in pool: pool.append(nm)
    return pool

def _same_gender_repl(removed, pool):
    if not pool: return None
    listing = "; ".join(pool[:30])
    try:
        ans = call_llm(f"Actor to match: {removed}. From this list, pick the FIRST actor of the SAME "
                       f"GENDER as {removed}. Reply with ONLY that one name, exactly as written.\n"
                       f"List: {listing}", max_tokens=16).strip()
        for p in pool:
            if p.lower() == ans.lower() or p.lower() in ans.lower() or ans.lower() in p.lower():
                return p
    except Exception:
        pass
    return pool[0]

def chat_pipeline(message, state):
    msg = message.lower(); genre = state.get("genre", ""); budget = state.get("budget")
    cast = state.get("cast", []); directors = state.get("directors", [])

    # ---- explicit "replace X with Y" (user names the replacement directly) ----
    old_name, new_name = _explicit_target(message)
    if new_name:
        retired = set(state.get("_retired", []))
        targets = _name_targets(cast, old_name.lower()) + _name_targets(directors, old_name.lower())
        if not targets:
            state["reply"] = f'I couldn\'t find "{old_name}" in the current cast or directors.'
            return state
        swapped = []
        for p in targets:
            for i, cc in enumerate(cast):
                if cc["name"] == p["name"]: cast[i] = {"name": new_name, "role": cc["role"]}; retired.add(p["name"])
            for i, dd in enumerate(directors):
                if dd["name"] == p["name"]: directors[i] = {"name": new_name, "role": dd["role"]}; retired.add(p["name"])
            swapped.append(f"{p['name']} → {new_name}")
        state["cast"] = cast; state["directors"] = directors; state["_retired"] = list(retired)
        state["reply"] = "Done — recast " + "; ".join(swapped) + "."
        return state

    rm_cast = _name_targets(cast, msg); rm_dirs = _name_targets(directors, msg)
    if (rm_cast or rm_dirs) and any(w in msg for w in SWAP_WORDS):       # ---- auto recast (gender-aware) ----
        retired = set(state.get("_retired", []))
        used = {c["name"] for c in cast} | {d["name"] for d in directors} | retired
        a_pool = [a for a in _people_pool(genre, budget, "actors") if a not in used]
        d_pool = [d for d in _people_pool(genre, budget, "directors") if d not in used]
        swapped = []
        for c in rm_cast:
            if not a_pool: break
            repl = _same_gender_repl(c["name"], a_pool)
            if not repl: break
            a_pool.remove(repl); used.add(repl); retired.add(c["name"])
            for i, cc in enumerate(cast):
                if cc["name"] == c["name"]: cast[i] = {"name": repl, "role": cc["role"]}
            swapped.append(f"{c['name']} → {repl}")
        for d in rm_dirs:
            if not d_pool: break
            repl = d_pool.pop(0); used.add(repl); retired.add(d["name"])
            for i, dd in enumerate(directors):
                if dd["name"] == d["name"]: directors[i] = {"name": repl, "role": dd["role"]}
            swapped.append(f"{d['name']} → {repl}")
        state["cast"] = cast; state["directors"] = directors; state["_retired"] = list(retired)
        state["reply"] = ("Done — recast " + "; ".join(swapped) + ".") if swapped else \
                         "I'm out of fresh ROI talent for that genre/budget tier."
        return state

    # ---- free-form edit: re-write the synopsis (premise-faithful), then re-judge ----
    cur = state.get("synopsis", ""); length = state.get("length") or 120
    prem = (state.get("premise") or "").strip()
    premise_block = f'\nThe revision MUST stay faithful to the original premise: "{prem}"\n' if prem else ""
    prompt = (f"You are a script doctor. Revise this {genre} film synopsis to satisfy the user's request.\n\n"
              f"Current synopsis:\n{cur}\n{premise_block}\nUser request: {message}\n\n"
              f"Rules:\n- Apply the request while keeping a coherent {genre} synopsis (~{length} words).\n"
              f"- Do not drop or contradict the original premise (if any).\n"
              f"- Output ONLY the revised synopsis prose. No preamble, no labels.")
    revised = _en(call_llm(prompt, max_tokens=min(900, int(length * 1.8) + 120)))

    v = validate_synopsis(revised, genre, state.get("window", ""), CONSTRAINTS, target_len=length)
    rounds = state.get("rounds", [])
    rounds.append({"n": len(rounds) + 1, "agent": "Refiner", "score": v["score"],
                   "failed": v["failed"], "excerpt": revised})
    state["synopsis"] = revised; state["score"] = v["score"]; state["rounds"] = rounds
    state["reply"] = f'Revised per your note: "{message}". New critic score {v["score"]:.2f}.'
    return state

class ChatReq(BaseModel):
    message: str
    state: dict

@api.post("/chat")
def chat(req: ChatReq):
    out = chat_pipeline(req.message, dict(req.state))
    return {"reply": out.pop("reply", ""), "result": out}

print("✅ /chat: explicit replace-with + gender-aware auto-recast + premise-faithful edits.")
````

````python
def _clean_name(s):
    return re.sub(r"[^A-Za-z0-9 .'-]", "", s or "").strip()

def _compare_chart(a, b):
    """Deterministic two-line median-ROI-per-year for two people (tries actors, then directors)."""
    for tbl, pcol in (("movie_actors", "actor"), ("directors", "director")):
        sql = (f"SELECT fin.{_YCOL} AS year, p.{pcol} AS person, "
               f"ROUND(MEDIAN(fin.roi),2) AS median_roi, COUNT(*) AS films "
               f"FROM fin JOIN {tbl} p ON fin.id = p.id "
               f"WHERE p.{pcol} ILIKE '%{a}%' OR p.{pcol} ILIKE '%{b}%' "
               f"GROUP BY year, p.{pcol} ORDER BY year LIMIT 200")
        try:
            cur = _acon.execute(sql); cols = [d[0] for d in cur.description]
            rows = [dict(zip(cols, r)) for r in cur.fetchmany(200)]
            if rows:
                return {"rows": rows, "columns": cols, "sql": sql,
                        "title": f"{a} vs {b} — median ROI per year",
                        "chart": "line", "x": "year", "y": "median_roi", "series": "person"}
        except Exception:
            continue
    return None

def ask_data(question):
    ylo, yhi = _YR
    # --- deterministic "compare A and B" / "A vs B" -> reliable two-line chart ---
    m = re.search(r"(?:compare\s+)?(.+?)\s+(?:and|vs\.?|versus|&)\s+(.+?)"
                  r"(?:\s+(?:over|by|across|through|during|in)\b.*)?$", question, re.I)
    if m and re.search(r"\bcompare\b|\bvs\.?\b|\bversus\b", question, re.I):
        a, b = _clean_name(m.group(1)), _clean_name(m.group(2))
        if a and b:
            res = _compare_chart(a, b)
            if res:
                return res
            return {"error": f"I couldn't find ROI data for “{a}” and/or “{b}”. "
                             "Check spelling, or try two well-known leads (e.g. DiCaprio and Brad Pitt)."}
    # --- text-to-SQL for everything else ---
    base = (f"You write ONE DuckDB SELECT. Tables: "
            f"fin(all movie columns incl. title, budget, revenue, {_YCOL}, roi), "
            f"movie_actors(id, actor), movie_genres(id, genre), directors(id, director). "
            f"RULES: 'actor' only in movie_actors, 'genre' only in movie_genres, 'director' only in directors; "
            f"join each via id = fin.id; filter people with ILIKE '%name%'. Use 'fin' for movie attributes; "
            f"no 'movies' table. Year col fin.{_YCOL} (data {ylo}-{yhi}). Always LIMIT 50.")
    def gen(extra=""):
        p = (base + extra + '\nReply ONLY JSON: {"sql":"<SELECT>","chart":"line|bar|none","x":"<col>",'
             '"y":"<col>","series":"<col or null>","title":"<short>"}\nQuestion: ' + question)
        return json.loads(call_llm(p, max_tokens=320, fmt="json"))
    try:
        spec = gen()
    except Exception:
        return {"error": "Couldn't turn that into a query — try rephrasing."}
    for attempt in range(2):
        sql = _safe(spec.get("sql", ""))
        if not sql:
            return {"error": "Generated query was unsafe/invalid.", "sql": spec.get("sql", "")}
        try:
            cur = _acon.execute(sql); cols = [d[0] for d in cur.description]
            rows = [dict(zip(cols, r)) for r in cur.fetchmany(50)]
            return {"rows": rows, "columns": cols, "sql": sql, "title": spec.get("title", ""),
                    "chart": spec.get("chart", "none"), "x": spec.get("x"), "y": spec.get("y"),
                    "series": spec.get("series")}
        except Exception as e:
            if attempt == 0:
                try:
                    spec = gen(f"\nPrevious SQL FAILED: {e}\nSQL: {sql}\nFix it (use 'fin', join people via id).")
                except Exception:
                    return {"error": f"Query failed: {e}", "sql": sql}
            else:
                return {"error": f"Query failed: {e}", "sql": sql}
````

/breakdown (cell 10c)

````python
# ---- /breakdown: budget allocation + ROI-weighted cast salary split (capped) ----
import statistics
from pydantic import BaseModel

_VFX_HEAVY = {"Science Fiction","Fantasy","Action","Adventure","Animation"}
_SPLIT_VFX = {"Cast (lead & supporting)":0.20,"Director, producers & writers":0.08,
              "Crew (below-the-line)":0.22,"Production (sets, locations, units)":0.16,
              "VFX & post-production":0.24,"Music & sound":0.05,"Contingency":0.05}
_SPLIT_STD = {"Cast (lead & supporting)":0.27,"Director, producers & writers":0.10,
              "Crew (below-the-line)":0.26,"Production (sets, locations, units)":0.20,
              "VFX & post-production":0.08,"Music & sound":0.05,"Contingency":0.04}

def _actor_roi(name):
    try:
        r = _acon.execute("SELECT ROUND(MEDIAN(f.roi),2), COUNT(*) FROM fin f "
                          "JOIN movie_actors ma ON f.id=ma.id WHERE ma.actor ILIKE ?",
                          ['%' + name + '%']).fetchone()
        if r and r[0] and r[1] and r[1] >= 2:
            return float(r[0])
    except Exception:
        pass
    return None

class BreakReq(BaseModel):
    budget: int
    genre: str = ""
    genres: list = []
    cast: list

@api.post("/breakdown")
def breakdown(req: BreakReq):
    gl = req.genres or ([req.genre] if req.genre else [])
    pct = _SPLIT_VFX if any(g in _VFX_HEAVY for g in gl) else _SPLIT_STD   # VFX-heavy if ANY genre is
    lines = [{"item":k,"pct":round(v*100),"amount":int(req.budget*v)} for k,v in pct.items()]
    cast_amt = int(req.budget * pct["Cast (lead & supporting)"])
    rois = [_actor_roi((c or {}).get("name","")) for c in req.cast]
    found = [x for x in rois if x]
    ref = statistics.median(found) if found else 1.0          # robust reference (median, not mean)
    billing = [0.45,0.27,0.18,0.10][:len(req.cast)]
    def _factor(roi):
        if not roi: return 1.0                                # off-pool / no record = neutral
        return min(max(roi/ref, 0.6), 1.6)                    # cap ROI swing to ±, keeps billing sane
    raw = [billing[i] * _factor(rois[i]) for i in range(len(req.cast))]
    s = sum(raw) or 1.0
    sal = [{"name":req.cast[i]["name"],"role":req.cast[i].get("role",""),
            "amount":int(cast_amt*raw[i]/s),"roi":rois[i]} for i in range(len(req.cast))]
    return {"breakdown":lines, "cast_salaries":sal, "cast_amount":cast_amt}

print("✅ /breakdown live (genre-list VFX + capped ROI-weighted cast split)")
````

````python
def _classify_genders(names):
    cache = globals().setdefault("_GENDER_CACHE", {})
    unknown = [n for n in dict.fromkeys(names) if n not in cache]
    if unknown:
        try:
            data = json.loads(call_llm(
                'Classify each actor by gender. Reply ONLY JSON {"M":[...],"F":[...]} using the EXACT '
                "names given.\nActors: " + ", ".join(unknown), max_tokens=300, fmt="json"))
            for g in ("M", "F"):
                for n in data.get(g, []):
                    for u in unknown:
                        if u.lower() == n.lower() or u.lower() in n.lower() or n.lower() in u.lower():
                            cache[u] = g
        except Exception:
            pass
        for u in unknown:
            cache.setdefault(u, "M")
    return cache

def _actor_of_gender(target, role, pool, used, gmap):
    for n in pool:
        if n not in used and gmap.get(n) == target:
            return n
    try:
        nm = call_llm(f"Name ONE well-known {'actress' if target == 'F' else 'actor'} who is "
                      f"age-appropriate to play '{role}'. Reply with ONLY the name.", max_tokens=12).strip()
        if nm and nm not in used:
            return nm
    except Exception:
        pass
    return None

def cast_for_script(synopsis, genre, budget):
    a_pool = _people_pool(genre, budget, "actors")[:14]
    d_pool = _people_pool(genre, budget, "directors")[:8]
    prompt = ("You are a casting director. Read the synopsis and cast ONLY characters who actually appear "
              "in it — do NOT invent new characters. ALWAYS include the protagonist and the characters most "
              "central to the conflict (the antagonist / those driving the danger or threat); pick the 3-4 "
              "most important and do NOT skip the main characters for minor ones.\n"
              f"Synopsis: {synopsis}\n"
              "For EACH character: note their gender AND approximate age from the story, then cast a "
              "well-known actor who matches BOTH (a child -> a child/teen actor; a young adult -> a young "
              "actor; middle-aged -> middle-aged; elderly -> an older actor). "
              f"Prefer these high-ROI {genre} actors: {', '.join(a_pool)}; if none fits the character's "
              "gender/age, pick ANY well-known actor who does.\n"
              'Reply ONLY JSON: {"cast":[{"actor":"<name>","character":"<role>","gender":"M|F"}],'
              '"directors":["<name>","<name>"]}.')
    try:
        data = json.loads(call_llm(prompt, max_tokens=340, fmt="json"))
        raw = [c for c in data.get("cast", []) if c.get("actor")][:4]
        dirs = [{"name": d, "role": "Director"} for d in data.get("directors", []) if d][:2]
    except Exception:
        raw, dirs = [], []
    if not raw:
        return [{"name": n, "role": "Cast"} for n in a_pool[:4]], (dirs or None)
    gmap = _classify_genders(a_pool + [c["actor"] for c in raw])
    cast, used = [], set()
    for c in raw:
        tgt = (c.get("gender") or "").upper()[:1]
        actor, role = c["actor"], c.get("character", "Cast")
        if tgt in ("M", "F") and gmap.get(actor) and gmap.get(actor) != tgt:
            actor = _actor_of_gender(tgt, role, a_pool, used, gmap) or actor
        used.add(actor); cast.append({"name": actor, "role": role})
    return cast, (dirs or None)

if not getattr(run_pipeline, "_cast_wrapped", False):
    _BASE_RUN = run_pipeline
def run_pipeline(text):
    res = _BASE_RUN(text)
    cast, dirs = cast_for_script(res["synopsis"], res["genre"], res.get("budget"))
    res["cast"] = cast
    if dirs:
        res["directors"] = dirs
    return res
run_pipeline._cast_wrapped = True
print("✅ casting: central characters + gender-verified + age-aware")
````
