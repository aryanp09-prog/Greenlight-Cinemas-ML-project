```python
import pandas as pd
import ast

def parse_genres(x):
    try:
        return [d['name'] for d in ast.literal_eval(x)] if isinstance(x, str) else []
    except (ValueError, SyntaxError):
        return []

def load_and_clean_movies(path):
    movies = pd.read_csv(path, low_memory=False)
    for col in ['budget', 'revenue', 'popularity', 'id']:
        movies[col] = pd.to_numeric(movies[col], errors='coerce')
    movies['release_date']  = pd.to_datetime(movies['release_date'], errors='coerce')
    movies['release_year']  = movies['release_date'].dt.year
    movies['release_month'] = movies['release_date'].dt.month
    movies['genre_list']    = movies['genres'].apply(parse_genres)
    return movies

movies = load_and_clean_movies('/content/data/movies_metadata.csv')
print("shape:", movies.shape)
fin = movies[(movies['budget'] > 0) & (movies['revenue'] > 0)].copy()
fin['roi'] = fin['revenue'] / fin['budget']
print(len(fin), "films usable for ROI")
```

```python
import pandas as pd

# Dedup movies on TMDB id (movies_metadata has a few duplicate rows) — demonstrates dedup
before = len(movies)
movies = movies.drop_duplicates(subset='id')
print(f"deduped movies: {before} -> {len(movies)} (removed {before - len(movies)})")

# IMDB official ratings: tconst, averageRating, numVotes (one row per title, already unique)
imdb = pd.read_csv('/content/data/title.ratings.tsv.gz', sep='\t')
print("imdb ratings shape:", imdb.shape)

# Direct join: movies.imdb_id ('tt0114709') == imdb.tconst ('tt0114709')
movies_imdb = movies.merge(imdb, left_on='imdb_id', right_on='tconst', how='left')

print("after join:", movies_imdb.shape)
print("movies with an IMDB rating:", movies_imdb['averageRating'].notna().sum())
print(movies_imdb[['title', 'vote_average', 'averageRating', 'numVotes']].head())
```

```python
import pandas as pd

recent = pd.read_csv('/content/data_recent/TMDB_movie_dataset_v11.csv', low_memory=False)
print("asaniczka raw:", recent.shape)

# Coerce numerics
for col in ['budget', 'revenue', 'popularity', 'id', 'vote_average', 'vote_count']:
    recent[col] = pd.to_numeric(recent[col], errors='coerce')

# Dates -> year/month
recent['release_date']  = pd.to_datetime(recent['release_date'], errors='coerce')
recent['release_year']  = recent['release_date'].dt.year
recent['release_month'] = recent['release_date'].dt.month

# NOTE: asaniczka 'genres' is a COMMA string ("Drama, Action"), not dicts like Rounak Banik
def parse_genres_csv(x):
    return [g.strip() for g in x.split(',')] if isinstance(x, str) and x.strip() else []
recent['genre_list'] = recent['genres'].apply(parse_genres_csv)

# Keep only NEW films: post-2017 AND not already in our set (dedup on id)
existing_ids = set(movies_imdb['id'].dropna())
new_films = recent[(recent['release_year'] > 2017) & (~recent['id'].isin(existing_ids))].copy()
print("new post-2017 films to add:", len(new_films))

del recent   # free ~2-3GB RAM

# Merge: union columns (cast/IMDB cols become NaN for new films — expected)
combined = pd.concat([movies_imdb, new_films], ignore_index=True)
combined = combined.drop_duplicates(subset='id')
print("combined shape:", combined.shape)

# Refresh the ROI subset on the bigger table
fin = combined[(combined['budget'] > 0) & (combined['revenue'] > 0)].copy()
fin['roi'] = fin['revenue'] / fin['budget']
print("ROI films now:", len(fin), "(was 5381)")
```

```python
!pip install -q duckdb
import os, duckdb

os.makedirs('/content/drive/MyDrive/greenlight',exist_ok = True)
DB_PATH = '/content/drive/MyDrive/greenlight/greenlight.duckdb'
con = duckdb.connect(DB_PATH)

# movies table: everything EXCEPT the python-list column (DuckDB prefers scalar cols)
movies_df = combined.drop(columns=['genre_list'])

# movie_genres: one row per (movie, genre) — the clean way to do genre SQL
mg_df = combined[['id', 'genre_list']].explode('genre_list').dropna(subset=['genre_list'])
mg_df = mg_df.rename(columns={'genre_list': 'genre'})

con.execute("CREATE OR REPLACE TABLE movies       AS SELECT * FROM movies_df")
con.execute("CREATE OR REPLACE TABLE movie_genres AS SELECT * FROM mg_df")

# fin as a VIEW — auto-derives the ROI subset, no duplicate storage
con.execute("""
CREATE OR REPLACE VIEW fin AS
SELECT *, revenue / budget AS roi
FROM movies
WHERE budget > 0 AND revenue > 0
""")

con.execute("CHECKPOINT")  # flush to disk

# verify
print("movies      :", con.execute("SELECT COUNT(*) FROM movies").fetchone()[0])
print("movie_genres:", con.execute("SELECT COUNT(*) FROM movie_genres").fetchone()[0])
print("fin (ROI)   :", con.execute("SELECT COUNT(*) FROM fin").fetchone()[0])
con.close()
print("saved to", DB_PATH)
```

```python
import pandas as pd, ast

credits = pd.read_csv('/content/data/credits.csv')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')

def get_director(crew_str):
    try:
        crew = ast.literal_eval(crew_str)
        names = [c['name'] for c in crew if c.get('job') == 'Director']
        return names[0] if names else None
    except (ValueError, SyntaxError):
        return None

credits['director'] = credits['crew'].apply(get_director)
directors_df = credits[['id', 'director']].dropna(subset=['director'])
print("films with a director:", len(directors_df))

con.execute("CREATE OR REPLACE TABLE directors AS SELECT * FROM directors_df")
con.execute("CHECKPOINT")
print("directors table saved")
```

```python
director_roi = con.execute("""
SELECT d.director,
       COUNT(*)                AS n_films,
       ROUND(MEDIAN(f.roi), 2) AS median_roi
FROM fin f
JOIN directors d ON f.id = d.id
GROUP BY d.director
HAVING COUNT(*) >= 5
ORDER BY median_roi DESC
LIMIT 15
""").df()
print(director_roi)
```

```python
def get_top_cast(cast_str, n=3):
    try:
        cast = ast.literal_eval(cast_str)
        return [c['name'] for c in cast[:n]]   # cast is ordered by billing
    except (ValueError, SyntaxError):
        return []

credits['top_cast'] = credits['cast'].apply(get_top_cast)
actors_df = credits[['id', 'top_cast']].explode('top_cast').dropna(subset=['top_cast'])
actors_df = actors_df.rename(columns={'top_cast': 'actor'})

con.execute("CREATE OR REPLACE TABLE movie_actors AS SELECT * FROM actors_df")
con.execute("CHECKPOINT")
print("movie_actors rows:", len(actors_df))
```

```python
actor_roi = con.execute("""
SELECT a.actor,
       COUNT(*)                AS n_films,
       ROUND(MEDIAN(f.roi), 2) AS median_roi
FROM fin f
JOIN movie_actors a ON f.id = a.id
GROUP BY a.actor
HAVING COUNT(*) >= 5
ORDER BY median_roi DESC
LIMIT 15
""").df()
print(actor_roi)
```

```python
import json, datetime

# --- Query the 4 insights ---
genres = con.execute("""
SELECT g.genre, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN movie_genres g ON f.id=g.id
GROUP BY g.genre HAVING COUNT(*)>=30
ORDER BY median_roi DESC LIMIT 5
""").df()
season = con.execute("""
SELECT release_month, COUNT(*) AS n_films, ROUND(MEDIAN(roi),2) AS median_roi
FROM fin WHERE release_month IS NOT NULL
GROUP BY release_month ORDER BY median_roi DESC
""").df()

directors = con.execute("""
SELECT d.director, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN directors d ON f.id=d.id
GROUP BY d.director HAVING COUNT(*)>=5
ORDER BY median_roi DESC LIMIT 10
""").df()

actors = con.execute("""
SELECT a.actor, COUNT(*) AS n_films, ROUND(MEDIAN(f.roi),2) AS median_roi
FROM fin f JOIN movie_actors a ON f.id=a.id
GROUP BY a.actor HAVING COUNT(*)>=5
ORDER BY median_roi DESC LIMIT 10
""").df()

# --- Assemble constraints (to_json/loads keeps it JSON-safe) ---
month_names = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
               7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
best_months  = season.head(3)['release_month'].astype(int).tolist()
worst_months = season.tail(2)['release_month'].astype(int).tolist()

constraints = {
    "meta": {
        "generated_on": datetime.date.today().isoformat(),
        "source": "TMDB (Rounak Banik + asaniczka) + IMDB ratings",
        "n_roi_films": int(con.execute("SELECT COUNT(*) FROM fin").fetchone()[0]),
        "budget_floor_usd": 100000,
        "roi_metric": "median of revenue/budget"
    },

    "top_genres": json.loads(genres.to_json(orient="records")),
    "seasonal_fit": {
        "best_months": best_months,
        "best_months_named": [month_names[m] for m in best_months],
        "worst_months_named": [month_names[m] for m in worst_months],
        "note": f"Summer ({month_names[best_months[0]]}/{month_names[best_months[1]]}) "
                f"and December strongest; September weakest."
    },
    "director_trends": json.loads(directors.to_json(orient="records")),
    "actor_trends": json.loads(actors.to_json(orient="records")),

    "audience_signal": {
  "loved_roi": 4.25, "mixed_roi": 1.74,
  "note": "Films audiences love (MovieLens >=3.8) earn ~2.5x the ROI of mixed films. Prioritize broad appeal."
}
}

# --- Save to Drive ---
OUT = '/content/drive/MyDrive/greenlight/constraints.json'
with open(OUT, 'w') as f:
    json.dump(constraints, f, indent=2, ensure_ascii=False)

print("saved:", OUT)
print(json.dumps(constraints, indent=2, ensure_ascii=False))

```

```python
# Aggregate 26M MovieLens ratings per film, map to TMDB id via links, persist
con.execute("""
CREATE OR REPLACE TABLE movielens AS
SELECT CAST(l.tmdbId AS BIGINT) AS id,
       ROUND(AVG(r.rating), 2)  AS ml_avg_rating,
       COUNT(*)                 AS ml_n_ratings
FROM read_csv_auto('/content/data/ratings.csv') r
JOIN read_csv_auto('/content/data/links.csv')   l ON r.movieId = l.movieId
WHERE l.tmdbId IS NOT NULL
GROUP BY l.tmdbId
""")
con.execute("CHECKPOINT")
print("films with MovieLens ratings:", con.execute("SELECT COUNT(*) FROM movielens").fetchone()[0])
print(con.execute("SELECT * FROM movielens ORDER BY ml_n_ratings DESC LIMIT 5").df())
```

```python
rating_vs_roi = con.execute("""
SELECT CASE WHEN ml.ml_avg_rating >= 3.8 THEN 'loved (>=3.8)'
            WHEN ml.ml_avg_rating >= 3.2 THEN 'liked (3.2-3.8)'
            ELSE 'mixed (<3.2)' END AS audience_band,
       COUNT(*)                AS n_films,
       ROUND(MEDIAN(f.roi), 2) AS median_roi
FROM fin f JOIN movielens ml ON f.id = ml.id
WHERE ml.ml_n_ratings >= 50
GROUP BY audience_band
ORDER BY median_roi DESC
""").df()
print(rating_vs_roi)
```
